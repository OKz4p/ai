정규표현식(regex)
```
    1. 참조 : https://ko.wikipedia.org/wiki/정규_표현식
    2. 간략문법
        \d : 숫자와 매치, [0-9]와 동일
        \D : 
        \w : 영문자나 숫자, [a-zA-Z0-9]와 동일 (한글이나 특수문자는 미포함)
        \W : 영문자나 숫자가 아닌 문자 [~!@#%^&*()_\-+]
        .  : 문자
        \. : (.)
        \s : ( )
        {n,} : n회 반복
        {m,n} : m번에서 n회 반복
        ? : (0번이나 1번 의미)
        + : (1번 이상 반복)
        * : (0번 이상 반복)
    3. 정규표현식 연습장 : https://regexr.com/
        ex1. 전화번호 패턴 : 0\d{1,2}[\-)\.\s]\d{3,4}[\-\s\.]\d{4} 
        ex2. 이메일 패턴 : \w[a-z0-9]+@\w+(\.\w+){1,2}
```

````python
data = "이름:홍길동, 주소:신림동, 전화:02-888-8888, \
핸드폰:010-9999-9999, 이메일:abc@lg.co.kr, abc@naver.com \
특징:서해번쩍 동해번쩍, 혈액형:B ㅋㅋ"
print(data)
````

````python
import re # re는 파이썬의 정규표현식 내장 라이브러리
tel_pattern = r"0\d{1,2}[\-)\.\s]\d{3,4}[\-\s\.]\d{4}"
tels = re.findall(tel_pattern, data) # findall 은 해당 패턴을 다 찾아줌. 없으면 빈 list([])
tels
````

````python
# 정규표현식 패턴과 처음 match되는 것을 찾아줘
firstTel = re.search(tel_pattern, data)
firstTel
````

````python
print(firstTel.group(0))
print(data[19:30])
````

````python
phone_pattern = r"(\d{2,3})-(\d{3,4})-(\d{4})"
phone = re.search(phone_pattern, data)
print(phone.group(0))
print(phone.group(1))
print(phone.group(2))
print(phone.group(3))
````

````python
email_pattern = r"([a-z0-9]+@\w+(?:\.\w+){1,2})"
print(re.findall(email_pattern, data))
print(re.search(email_pattern, data).group(0))
````

````python
# 특정 문자외 없애기
re.sub(r"[^가-힣0-9\-a-z@]"," ", data) # ^은 반대..
````
