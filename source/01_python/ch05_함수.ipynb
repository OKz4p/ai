## 지역변수(local var)와 전역변수(global var)

````python
global_var = 100 # 전역변수
def func1():
    print("함수안 :", global_var)
func1()
print("함수밖 :", global_var)
````

````python
def func2():
    local_var = 200 # 지역변수
    print("함수안 :", local_var)
func2()
print("함수밖 :", local_var)
````

````python
# 변수의 값을 조회하는 순서 : 지역변수(지역심볼테이블) > 전역변수(전역심볼테이블) > 내장된 이름
g_var = 100 # 전역변수
def func3():
    g_var = 200 # 지역변수
    print("함수안 g_var:", g_var)
func3()
print("함수밖 g_var :", g_var)
````

````python
# 렉시컬 특징 : 변수가 어디에 선언(할당)되었는지에따라 변수 범위가 정해짐
g_var = 100 # 전역변수
def func3():
    global g_var # func3 함수내의 g_var는 전역변수를 쓴다
    print("함수안 g_var:", g_var) # 전역변수
    g_var = 200 # 전역변수
func3()
print("함수밖 g_var :", g_var)
````

````python
# 함수 정의
def add(a, b):
    return a+b
def add(a, b, c):
    return a+b+c
````

````python
# args 앞의 * : 매개변수들(인수들)이 args 튜플로 전달
def add(*args):
    "전달받은 매개변수들의 합을 return. 매개변수가 없을 겨우 0을 return"
    # print(args) # (1, 2, 3)
    tot = 0 # 누적변수
    for num in args:
        tot += num # tot = tot+num
    return tot
````

````python
print(add())
print(add(1))
print(add(10, 20))
print(add(10, 2, 3))
print(add(1, 2, 3, 4, 5))
````

## 인수의 언패킹
### 튜플인수 언패킹
- 패킹 : 함수 정의시 매개변수 앞에 *붙임(튜플매개변수)
- 언패킹 : 함수 호출시 매개변수 앞에 *붙임(리스트나 튜플이 각자 요소별 매개변수로)

````python
def add(*args):
    tot = 0 # 누적변수
    for num in args:
        tot += num 
    return tot
````

````python
numbers = [1, 2, 3, 4, 5]
add(*numbers)
````

````python
def map_template(func, L=[]):
    result=[]
    for item in L:
        result.append(func(item))
    return result
````

### (2) filter

````python
def filter_template(func, L):
    result = []
    for item in L:
        if func(item):
            result.append(item)
    return result
````

````python
def even_chk(value): # 짝수면 True return, 홀수면 False를 return
    return value%2==0
even_chk = lambda value : value%2==0
even_chk(9)
````

````python
list_data = [1, 2, 3, 4, 5, 6, 7, 8, 9]
filter_template(lambda value: value%2==0, list_data)
````

````python
[value for value in list_data if value%2==0]
````

````python
list(filter(lambda value : value%2==0, list_data))
````

````python
# 4 함수 인자로 시작과 끝 숫자가 주어질 때  시작부터 끝까지의 모든 정수값의 합을 리턴하는 함수를 작성하세요(시작값과 끝값포함)
def mysum(fRom, end):
    numsum = 0
    for num in range(fRom, end+1):
        numsum += num
    return numsum

print(mysum(1, 3))
````

````python
# 5 함수의 인자로 문자열을 포함하는 리스트가 입력될 때 각 문자열의 첫 세글자로만 구성된 리스트를 리턴하는 함수와 람다식을 작성하세요
#예를 들어 함수의 입력으로 [‘Seoul’, ‘Daegu’, ‘Kwangju’, 'Jeju']가 입력될 때 함수의 리턴값은 [‘Seo’, ‘Dae’, 'Kwa', ‘ Jej ’]
def get_abbrs(lst):
    return [s[:3] for s in lst]

city = ['Seoul', 'Daegu', 'Kwangju', 'Jeju']
print(get_abbrs(city))  

abbrs= lambda lst : list(map(lambda s: s[:3], lst))
print(abbrs(city))  
````

````python
# 10 n개의 벡터(리스트나 튜플, 셋등)들의 크기가 같은지 여부를 reutrn하는
# vector_size_check(*vector_var)를 한줄의 코드(리스트 컴프리헨션 이용)로 작성하시오
#def vector_size_check(*vector_var):
#    print(vector_var)
#    return False


def vector_size_check(*vector_var):
    return len(set([len(v) for v in vector_var])) ==1 
print(vector_size_check([1], ("a","b"), {"name":"홍", "age":22},{1,2}))
print(vector_size_check([1, 2], ("a","b"), {"name":"홍", "age":22},{1,2}))
````
