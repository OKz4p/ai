## 지역변수(local var)와 전역변수(global var)

````python
global_var = 100 # 전역변수
def func1():
    print("함수안 :", global_var)
func1()
print("함수밖 :", global_var)
````

````python
def func2():
    local_var = 200 # 지역변수
    print("함수안 :", local_var)
func2()
print("함수밖 :", local_var)
````

````python
# 변수의 값을 조회하는 순서 : 지역변수(지역심볼테이블) > 전역변수(전역심볼테이블) > 내장된 이름
g_var = 100 # 전역변수
def func3():
    g_var = 200 # 지역변수
    print("함수안 g_var:", g_var)
func3()
print("함수밖 g_var :", g_var)
````

````python
# 렉시컬 특징 : 변수가 어디에 선언(할당)되었는지에따라 변수 범위가 정해짐
g_var = 100 # 전역변수
def func3():
    global g_var # func3 함수내의 g_var는 전역변수를 쓴다
    print("함수안 g_var:", g_var) # 전역변수
    g_var = 200 # 전역변수
func3()
print("함수밖 g_var :", g_var)
````

````python
# 함수 정의
def add(a, b):
    return a+b
def add(a, b, c):
    return a+b+c
````

````python
# args 앞의 * : 매개변수들(인수들)이 args 튜플로 전달
def add(*args):
    "전달받은 매개변수들의 합을 return. 매개변수가 없을 겨우 0을 return"
    # print(args) # (1, 2, 3)
    tot = 0 # 누적변수
    for num in args:
        tot += num # tot = tot+num
    return tot
````

````python
print(add())
print(add(1))
print(add(10, 20))
print(add(10, 2, 3))
print(add(1, 2, 3, 4, 5))
````

## 인수의 언패킹
### 튜플인수 언패킹
- 패킹 : 함수 정의시 매개변수 앞에 *붙임(튜플매개변수)
- 언패킹 : 함수 호출시 매개변수 앞에 *붙임(리스트나 튜플이 각자 요소별 매개변수로)

````python
def add(*args):
    tot = 0 # 누적변수
    for num in args:
        tot += num 
    return tot
````

````python
numbers = [1, 2, 3, 4, 5]
add(*numbers)
````

````python
def map_template(func, L=[]):
    result=[]
    for item in L:
        result.append(func(item))
    return result
````
