# 피클을 이요한 객체 저장 및 불러오기
## 형식이 있는 텍스트 데이터 불러오기
- ch09_member.txt => 객체 list 

```python
with open('data/ch09_member.txt', 'r', encoding='utf-8') as f:
    lines = f.readlines()
for line in lines:
    #print(line)
    data = line.strip().split(',')
    name = data[0]
    age = int(data[1])
    email = data[2]
    address = data[3]
    print(f"이름:{name}, 나이:{age}, 메일:{email}, 주소:{address}")
```

```python
class Member:
    def __init__(self, name, age, email, address):
        self.name = name
        self.age = age
        self.email = email
        self.address = address
    def __str__(self):
        return f"{self.name},{self.age},{self.email},{self.address}"
    def __eq__(self, other):
        if isinstance(other, Member):
            return self.__str__() == other.__str__()
        return False
    
    def as_dict(self): # 객체를 딕셔너리로 리턴
        return {"name":self.name, 'age':self.age, 'email':self.email, 'address':self.address}
    def as_list(self): # 객체를 list로 리턴
        return [self.name, self.age, self.email, self.address]
```

```python
member = Member("홍길동", 33, 'hong@h.co.kr', '신림동')
member2 = Member("홍길동", 33, 'hong@h.co.kr', '신림동')
print(id(member), id(member2))
print(member == member2)
print(member.__eq__(member2))
print(member)
print(member.as_dict())
print(member.__dict__)
print(member.as_list())
```

```python
# 파일의 내용을 member list나 딕셔너리 list로 만들기
member_list = [] # 파일의 내용이 member객체로 append
member_dict_list = []
with open('data/ch09_member.txt', 'r', encoding='utf-8') as f:
    lines = f.readlines()
for line in lines:
    #print(line)
    data = line.strip().split(', ')
    name = data[0]
    age = int(data[1])
    email = data[2]
    address = data[3]
    member = Member(name, age, email, address)
    #print(member)
    member_list.append(member)
    member_dict_list.append(member.__dict__)
```

## 피클링

```python
# 피클링(확장자 : pkl, pickle, data)을 이용한 객체를 파일에 저장하기
import pickle
with open('data/ch09_member.pkl', 'wb') as f:
    pickle.dump(member_list, f)
```

```python
# 피클링을 이용한 파일에서 객체 데이터 읽기
with open('data/ch09_member.pkl', 'rb') as f:
    loaded_member_list = pickle.load(f)
```

```python
# member_list
# loaded_member_list
for member, loaded_member in zip(member_list, loaded_member_list):
    print(member, '로드된 member=', loaded_member)
```
