```python
def safe_index(lst, item):
    '''
    첫번째 매개변수 lst에서 item 요소가 있는 index를 반환하는 함수
    item 요소가 없으면 -1를 반환
    '''
    if item in lst:
        return lst.index(item)
    else:
        return -1
```

```python
def safe_remove(lst, remove_item):
    '''
    첫번째 매개변수 lst에서 remove_item 데이터 제거
    해당 remove_item이 lst에 없으면 메세지 출력
    '''
    if remove_item in lst:
        lst.remove(remove_item)
    else:
        print("해당 데이터는 없는 데이터입니다")
```

```python
names = ["길동", "길자", "순자"]
tels = ["010-9999-9999",
       "010-8888-8888",
       "010-7777-7777"]
adds = ["서울", "인천","판교"]
for idx in range(len(names)): # [0,1,2]
    print("{}. {}의 연락처는 {}이고, {}에 살아요".format(idx+1,
                                             names[idx],
                                             tels[idx],
                                             adds[idx]))
```

```python
list(zip(names, tels, adds))
```

```python
for name, tel, add in zip(names, tels, adds):
    print(f"{name}의 연락처는 {tel}이고, {add}에 살아요")
```

```python
list(enumerate(zip(names, tels, adds)))
```

```python
for idx, (name, tel, add) in enumerate(zip(names, tels, adds)):
    print(f"{idx+1}. {name}의 연락처는 {tel}이고, {add}에 살아요")
```

```python
# names, tels, add를 딕셔너리 리스트 or 이차원
members = []
for name, tel, add in zip(names, tels, adds):
    member = {"name":name, "tel":tel, "address":add}
    members.append(member)
members
```

```python
members = []
for idx, (name, tel, add) in enumerate(zip(names, tels, adds)):
    member = {"순위":idx+1, "이름":name, "전화":tel, "주소":add}
    members.append(member)
members
```
