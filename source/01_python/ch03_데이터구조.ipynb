## 인덱싱과 요소 인덱스 찾기
- 변수[idx] : 인덱스를 이용한 접근(없는 인덱스에 사용시 에러)
- count(찾을데이터) : 리스트에서 해당 찾을 데이터가 나오는 갯수 반환
- index(찾을데이터) : 0번째부터 해당 찾을데이터가 몇번째 index에 있는지 반환
- index(찾을데이터, start) : start번째부터 해당 찾을 데이터가 몇번째 index에 있는지 반환 

```python
numbers = [1, 2, 3, [1, 2, 3], 1, 2, 3]
```

```python
numbers.count(2)
```

```python
numbers.count(9) # 없는 데이터를 넣어도 오류가 안뜨고 대신 0으로 출력한다
```

```python
if 9 in numbers:
    print(numbers.index(9))
else:
    print("찾을 수 없음")
```

```python
def safe_index(lst, item):
    '''
    첫번째 매개변수 lst에서 item 요소가 있는 index를 반환하는 함수
    item 요소가 없으면 -1를 반환
    '''
    if item in lst:
        return lst.index(item)
    else:
        return -1
```

```python
from ch03 import safe_index # ch03모듈안의 safe_index 로드
numbers = [1, 2, 3]
safe_index(numbers, 2) # numbers가 lst, item은 2로 할당 참이면 if문, 거짓이면 -1이 나옴
```

## 원하는 요소 추출

1. 리스트 슬라이싱 : list에서 원하는 index의 데이터들을 추출(인덱스 범위를 벗어나도 에러나지 않음)
    -[from, stop, step] : from부터 stop앞까지 step씩 전진하면서 index 값 추출
       * from 생략시 처음, stop생략시 끝, step생략시 1. 모두 생략가능
       * 인덱스 음수 가능
        
2. <font color="red">리스트 컴프리헨션</font> : list에서 원하는 조건의 데이터로 리스트 생성

```python
numbers = [1, 3, 5, 9, 10, 12, 13, 18]
# 거꾸로 출력, 짝수번째 출력, 다 출력(슬라이싱 이용)
print(numbers[::-1])
print(numbers[1::2])
print(numbers[::])    
```

```python
# 홀수만 추출
# lst = []
# for item in numbers:
#     if item%2:
#         lst.append(item)
lst = [item for item in numbers if item%2]
lst
```

```python
# 짝수만 추출
numbers = [1, 3, 5, 9, 10, 12, 13, 18]
lst = [item for item in numbers if item%2 == 0]
lst
```

```python
# numbers에서 10이상만 추출
[n for n in numbers if n>=10]
```

```python
# 과일의 글자가 5글자 이상인 과일만 추출
fruits = ["orange", "apple", "mango", "kiwi"]
fruits_5 = [fruit for fruit in fruits if len(fruit) >=5]
print(fruits_5)
```

```python
# fruits 리스트에서 과일 글자수를 추출
fruits_word = [len(w) for w in fruits]
print(fruits_word)
```

```python
# lst에 3이상인 요소들의 제곱을 추출
lst = [1, 2, 3, 4, 5]
res = [x**2 for x in lst if x>3]
print(res)
```

```python
# lst의 두배값을 추출
ress = [y**2 for y in lst]
print(ress)
```

## 요소 삭제하기

```python
# pop()    : 가장 마지막 인덱스의 요소를 반환하고 삭제
# pop(idx) : idx번째 요소를 반환하고 삭제
# del 리스트변수[idx] : idx번째 요소를 삭제. 잘안쓴다네요
# remove(data) : list안의 data를 삭제 (data가 요소로 없으면 에러) =>있는지 확인을 하고 사용해야함(in연산자 or count()사용)
```

```python
numbers = [1, 2, 3, 4, 5]
print("삭제한 데이터 :", numbers.pop())
print("삭제후 numbers :", numbers)
print("삭제한 데이터 :", numbers.pop())
print("삭제후 numbers :", numbers)
print("삭제한 데이터 :", numbers.pop(0))
print("삭제후 numbers :", numbers)
del numbers[0]
print("삭제후 numbers :", numbers)
```

```python
def safe_remove(lst, remove_item):
    '''
    첫번째 매개변수 lst에서 remove_item 데이터 제거
    해당 remove_item이 lst에 없으면 메세지 출력
    '''
    if remove_item in lst:
        lst.remove(remove_item)
    else:
        print("해당 데이터는 없는 데이터입니다")
```

## 리스트 복제하기

```python
# 스칼라 변수의 복제
i = 10
copied_i = i
print(f"원본i={i}, 복사본i={copied_i} / 원본주소 : {id(i)}, 복사본주소 : {id(copied_i)}")
copied_i += 5 # copied_i = copied_i + 5
print(f"원본i={i}, 복사본i={copied_i} / 원본주소 : {id(i)}, 복사본주소 : {id(copied_i)}")
```

```python
# 데이터 구조의 복제
numbers = [5, 1, 9] # 원본
new_numbers = numbers # 복제본
print(f"원본i={numbers}, 복사본i={new_numbers} / 원본주소 : {id(numbers)}, 복사본주소 : {id(new_numbers)}")
new_numbers[0] = 55
print(f"원본i={numbers}, 복사본i={new_numbers} / 원본주소 : {id(numbers)}, 복사본주소 : {id(new_numbers)}")
```

```python
# 복제(=깊은 복사)
numbers = [5, 1, 9] # 원본
new_numbers = numbers.copy()
print(f"원본i={numbers}, 복사본i={new_numbers} / 원본주소 : {id(numbers)}, 복사본주소 : {id(new_numbers)}")
new_numbers.sort()
print(f"원본i={numbers}, 복사본i={new_numbers} / 원본주소 : {id(numbers)}, 복사본주소 : {id(new_numbers)}")
```
