# 1절. 조건문(if)
```
if 조건1:
    조건1이 참일 경우 수행할 명령어들(if문 단독 가능. 들여쓰기 중요)
elif 조건2:
    조건1이 거짓이고, 조건이2가 참일 경우 수행할 명령어들(elif는 필요한만큼 사용가능)
else:
    조건이 만족하지 않을 때 수행할 명령어들(else는 단독사용불가)
```

````python
score = int(input("Enter you score >"))
if score>=60:
    print("pass입니다")
else:
    print("fail입니다")
````

## for-in문

````python
for i in range(1, 10, 2): # [1, 3, 5, 7, 9]
    print(i, end="\t")
else:
    print("\nDONE")
    print("END")
````

## while문
```
while 조건:
    조건이 참일 경우 반복할 문장들
else:
    조건이 거짓일 경우 한번 실행할 문장(else 선택사항)
```

````python
# 1~100까지 짝수들의 합은 계산하고 출력하시오(while문)
tot = 0 
i = 1
while i<=100:
    if i%2==0:
        tot += i # 여기까지만 적어 실행하면 무한반복문
    i += 1 # i변수 1증가
else:
    print("1~100까지 짝수의 합은", tot)
````

````python
# 사용자에게 두 정수를 입력받아 작은수부터 큰수까지 홀수들의 합을 누적하여 출력하시오
first = int(input("첫번째 정수는?"))
second = int(input("두번째 정수는?"))
# (1) 두수 중 작은 수는 minNo, 큰수는 maxNO에 할당
if first<second:
    minNo = first
    maxNo = second
else:
    minNo = second
    maxNo = first
# minNo = first if first<second else second
# maxNo = first if first>second else second 
# (2) minNo~maxNo까지 홀수 누적하기
tot = 0
for no in range(minNo, maxNo+1):
    if no%2: # if no%2!=0
        tot += no
print(f"{minNo}부터 {maxNo}까지 홀수의 합은 {tot}입니다")
````

````python
# ex1 사용자에게 두 정수를 입력받아 작은수부터 큰수까지 홀수들의 합을 누적하여 출력하시오(pdf 9p.)
first = int(input("첫번째 정수는?"))
second = int(input("두번째 정수는?"))
# (1) first에 작은 수를 second에 큰수
if first>second:
    # 변수 바꾸는 방법
#    temp = first
#    first = second
#    second = temp
    first, second = second, first # 동시에 변수를 바꾸는 파이썬에서만 가능한 일
# (2) first부터 second까지홀수합 누적하기
tot = 0
for no in range(first, second):
    if no%2: # if no%2!=0
        tot += no
print(f"{first}부터 {second}까지 홀수의 합은 {tot}입니다")
````

````python
# 난수 발생하기 
import random
#random.seed(5) 난수를 일치시키기 위한 seed값 설정(설정을 날리려면 커널 재시작)
print(random.random()) # 0부터 1미만의 실수 난수(임의의 수)
print(random.randint(1, 45)) # 1부터 45까지의 정수 난수(임의의 수)
print(random.choice(range(1, 46))) # 1부터 45까지의 정수 난수
print(random.choice(["남자","여자"]))
print(random.choices(["남자","여자"], weights=[0.75, 0.25])) # 남자나 여자 중 하나 선택(남자가 선택될 확률이 0.75%)
print(random.choices(["남성","여성","중성"], k=2)) # k개 복원추출
print(random.choices(range(1, 46), k=6))
print(random.sample(range(1, 46), k=6)) # k개 비복원추출
````

````python
# ex, 1~100사이의 정수를 맞추는 프로그램을 구현하시오
'''
(프로그램 안에서 random수가 만들어지고 그 random수 맞추기)
1단계 : 수를 못맞추면 UP/DOWN 메시지를 출력하고 숫자를 맞추면 축하 메시지 뿌리고 끝남(1-100사이의 수를 알아맞추기)
2단계 : 수를 못맞추면 min~max를 조정, 맞추면 축하메시지 뿌리고 끝남
3단계 : 2단계의 기회를 5회로 제한. 5회의 기회가 지나면 실패메시지 뿌리고 끝남
(2단계와 3단계는 연습문제로)
'''
import random
miN = 1
maX = 100
randomNum = random.randint(1, 100)
print("컴퓨터가 발생한 임의의 수 :", randomNum)
inputNum = 0
miN = 1
maX = 100
for cnt in range(5,0,-1): # while inputNum != randomNum: (1,2단계)
    inputNum = int(input(f"{miN}~{maX}사이의 수를 맞춰보세요(남은기회:{cnt}번) :"))
    if (inputNum>maX) | (inputNum<miN):
        print("수의 범위를 벗어났습니다")
    elif inputNum>randomNum:
        print("DOWN")
        maX = inputNum -1
        # 2단계는 max값 조정
    elif inputNum<randomNum:
        print("UP")
        miN = inputNum + 1
    else:
        print("ㅊㅋㅊㅋ")
        break
else:
    print("기회를 모두 사용하셨습니다. 실패입니다")
print("DONE")
````

````python
# 중첩 루프 탈출 방법 1 : flag변수 이용
flag = False
for a in range(0, 3): # a <= 0, 1, 2
    for b in range(1, 3): # b <= 1, 2
        if a == b:
            print(f"a={a}, b={b} break")
            flag = True
            break
        print(a, b)
    if flag:
        break
print("DOWN")
````
