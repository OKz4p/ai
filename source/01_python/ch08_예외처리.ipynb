## 예외를 지정한 처리
```
try:
    예외가 발생할 수도 있는 문장들
except 예외타입1:
    해당 예외가 발생할 경우 실해할 문장
except 예외타입2:
    해당 예외가 발생할 경우 실해할 문장
```
**여러 예외처리 할 경우 상위 예외타입은 아래로**

````python
# 100을 나눌 유효한 정수를 입력할 때까지 정수를 입력받아 유효한 정수를 입력하면 입력한 정수와 100으로 나눈 수를 출력
# ZeroDivisionError : "division by zero"
# ValueError : invalid literal for int() with base 10: 'a'
while True:
    try:
        x = int(input("100을 나눌 정수를 입력하세요 > "))
        print(f"입력한 정수는 {x}")
        print("100을 입력한 정수로 나누면 {:.2f}".format(100/x))
        break 
        
    except ZeroDivisionError:
        print("0으로 나눌 수 없습니다. 다시 입력해주세요.")
    except ValueError:
        print("정수가 아닙니다. 다시 입력하세요.")
    except Exception:
        print("원인 모를 예외가 발생했습니다. 다시 입력하세요.")
print("DONE")
````

````python
# 100을 나눌 유효한 정수를 입력할 때까지 정수를 입력받아 유효한 정수를 입력하면 입력한 정수와 100으로 나눈 수를 출력
# ZeroDivisionError : "division by zero"
# ValueError : invalid literal for int() with base 10: 'a'
while True:
    try:
        x = int(input("100을 나눌 정수를 입력하세요 > "))
        print(f"입력한 정수는 {x}")
        print("100을 입력한 정수로 나누면 {:.2f}".format(100/x))
        break 
        
    except (ValueError, ZeroDivisionError): # 예외 타입을 튜플처럼 하나로 작성 가능
        print("100으로 나눌 수 있는 정수가 아닙니다. 다시 입력하세요.")
    except Exception: # 상위 예외는 아래로
        print("원인 모를 예외가 발생했습니다. 다시 입력하세요.")
print("DONE")
````

## 예외인수(e)
- e를 변수로 안받아도 ㄱㅊ은데 대부분 e를 사용한다네요
```
try:
```

```
try:
    예외가 일어날 수도 있는 명령어들
except [예외타입 [as e]]:
    해당 예외타입시 수행할 명령어
[else: # 잘 사용하지 않음
    try블록을 다 수행한 후 실행할 명령어]
finally:
    try블록이나 except블록을 수행한 후 실행
```

````python
try:
    f = open("data/ch08.txt", "r") # encoding='utf-8) # FileNotFoundError
    data = f.read() # UnicodeDecodeError
except (FileNotFoundError, UnicodeDecodeError) as e: # 예외를 모르겠다면 상위예외인 Exception 사용가능
    print(e)
else:
    print(data) # try불록을 다 실행한 후 실행
finally:
    f.close()
````

````python
# 사용자 정의 예외 : Exception클래스를 상속받은 클래스
class LengthZeroError(Exception):
    '매개변수 길이가 0일때 발생하는 예외'
    def __init__(self, message):
        super().__init__(message) # 부모클래스의 생성자에서 예외메세지를 세팅
````

````python
def args_print(*data): # 튜플매개변수
    if len(data)==0:
        raise LengthZeroError("매개변수 길이가 0이면 예외발생(사용자 정의 예외)")
    for item in data:
        print(item, end=' ')
    print("등을 입력하셨습니다")
````

````python
args_print(1, 2, 4)
````

````python
args_print()
````

````python
try:
    args_print()
except Exception as e:
    print(e)
````

# 추상클래스
- 추상클래스 : 추상메소드가 1개 이상 포함된 객체를 생성할 수 없는 클래스
- 추상메소드 : 호출할 수 없는 메소드. 상속받은 클래스에서 재정의를 강요할 때 사용
- 파이썬에서의 추상클래스는 raise를 이용하거나, @abstractmethod을 이용하여 추상메소드를 구현

````python
class Shape: # 추상클래스
    def __init__(self, x, y):
        raise NotImplementedError('객체를 생성할 수 없음')
        pass
    def calc_area(self):
        raise NotImplementedError('추상메소드')
````

````python
from abc import abstractmethod, ABC
class Shape(ABC):
    @abstractmethod
    def __init__(self):
        pass
    @abstractmethod
    def calc_area(self):
        pass
````

````python
s = Shape()
````

````python
class Circle(Shape):
    def __init__(self, r=3):
        self.r = r
    def calc_area(self):
        import numpy as np
        return np.pi * (self.r**2)
````

````python
myCircle = Circle(5)
myCircle.calc_area()
````

# 파일 정리 작업(with절 이용)

````python
try:
    f = open("data/test.txt", "r", encoding = 'cp949') # windows에서 기본값 cp949 / 리눅스나 맥ox는 utf-8
    lines = f.readlines() # 한줄한줄 모든 줄을 읽어 list로 return
    print(lines)
except (FileNotFoundError, UnicodeDecodeError) as e:
    print(e)
finally:
    f.close() # 제일 중요 여기서는 에러가 안나지만 다른 곳에서 무한반복문처럼 돌고 끝남(=에러가 일어난다는뜼)
````

````python
# with절 이후에는 자동적으로 close() 실행
# 파일 접근시 finally 대신 이렇게 with절을 많이 사용한다
try:
    with open("data/test.txt", "r") as f:
        lines = f.readlines()
    print(lines)
except (FileNotFoundError, UnicodeDecodeError) as e:
    print(e)
````
