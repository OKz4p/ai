# try ~ except로 예외처리
## try ~ except
```
try:
    예외가 발생할 가능성이 있는 명령어들
except:
    예외가 발생했을 경우 실행할 명령어들
```

````python
# 100을 입력받은 정수값으로 나눠 출력
x = int(input("정수를 입력하세요 >"))
print(f"입력한 정수는 {x}")
print("100을 입력한 정수로 나누면 {:.2f}".format(100/x))
# ZeroDivisionError : "division by zero"
#                    0을 입력했을때 예외 타입
# ValueError : invalid literal for int() with base 10: 'a'
#              invalid literal(잘못된 변수)를 입력했을때 예외타입     
````

````python
# 100을 입력받은 정수값으로 나눠 출력
x = int(input("정수를 입력하세요 >"))
print(f"입력한 정수는 {x}")
print("100을 입력한 정수로 나누면 {:.2f}".format(100/x))
# ZeroDivisionError : "division by zero"
#                    0을 입력했을때 예외 타입
# ValueError : invalid literal for int() with base 10: 'a'
#              invalid literal(잘못된 문자)를 입력했을때 예외타입     
````

````python
# 100을 입력받은 정수값으로 나눠 출력
# ZeroDivisionError : "division by zero"
# ValueError : invalid literal for int() with base 10: 'a'
try:
    x = int(input("정수를 입력하세요 >"))
    print(f"입력한 정수는 {x}")
    print("100을 입력한 정수로 나누면 {:.2f}".format(100/x))
except:
    print("나누기에 유효한 정수가 아닙니다")
print("DONE")
````

````python
while True:
    try:
        x = int(input("정수를 입력하세요 > "))
        print(f"입력한 정수는 {x}")
        print("100을 입력한 정수로 나누면 {:.2f}".format(100/x))
        break 
        
    except (ValueError, ZeroDivisionError):
        print("나누기에 유효한 정수가 아닙니다. 다시 입력해주세요.")
print("DONE")
````
